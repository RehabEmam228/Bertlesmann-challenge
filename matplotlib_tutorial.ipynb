```
import matplotlib.pyplot as plt
```

#**Linear** **Plot**

```
x = [1,2,7]
y = [4,8,2]
x2 = [10,9,2]
y2 = [6,11,12]

plt.plot(x, y, label='First line', linewidth=5) #linewidth to change thickness of the line
plt.plot(x2, y2, label='Second line')

plt.title('Interesting Graph\nCheck it out')
plt.xlabel('Probabilities')
plt.ylabel('Classes')

#legend is responsible of the rectangle showing(first & second lines)
plt.legend()
plt.show()
```

#**Bar** **Plot**

```
x = [1,4,7,8,10]
y = [6,9,10,15,7]
x2 = [3,5,2,9]
y2 = [2,4,1,9]

plt.bar(x, y, label='First bar', color='pink')
plt.bar(x2, y2, label='Second bar', color='green')

plt.legend()
plt.show()
```

#**Scatter** **Plot**

```
x = [2,5,3,7]
y = [1,3,6,8]
z = [5,7,8,11]
plt.scatter(x, y, marker='o', s=50, label='age')
plt.scatter(x, z, marker='*', s=50, label='disease')
plt.legend()
plt.show()
```

There is also plt.stackplot

```
efficiency = [7,9,11,6]
probability = [8,2,6,5]
durability = [11,5,3,9]
loading = [6,2,3,5]

slices = [3,4,5,2] # slicing the pie
editors = [efficiency, probability, durability, loading]
color = ['cyan','red','blue','black']

plt.pie(slices, labels=editors, colors=color)
plt.show()
```
