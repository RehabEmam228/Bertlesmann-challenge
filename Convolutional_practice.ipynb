# Read then show image

```
img = mpimg.imread("/content/drive/My Drive/building_landscape.jpg")
plt.imshow(img)
```

# Change image to gray scale

```
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap="gray")
```

# Filters

1. Create a Sobel kernel to detect edges

```
# filter to detect horizontal edges
horizontal_sobel = np.array([[-1, -1.5, -2, -1.5, -1],
                    [-0.5, -1, -1.5, -1, -0.5],
                    [0, 0, 0, 0, 0],
                    [0.5, 1, 1.5, 1, 0.5],
                    [1, 1.5, 2, 1.5, 1]]) 
                   

```

```
# filter to detect vertical edges
vertical_sobel = np.array([[-1, 0, 1],
                           [-1, 0, 1],
                           [-1, 0, 1]])
```



 2. Sobel to blur image



```
sobel_x = cv2.blur(img, (10, 10))
# gaussian_blur = cv2.GaussianBlur(img, (5, 5), 0)
# median_blur = cv2.medianBlur(img, 5)
```

# Apply filters

# If I use "vertical_sobel" which detects vertical edges

```
# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(img_gray, -1, vertical_sobel)

plt.imshow(filtered_image, cmap='gray')
```

# If I use horizontal_sobel which detects horizontal edges

```
# Filter the image using filter2D, which has inputs: (grayscale image, bit-depth, kernel)  
filtered_image = cv2.filter2D(img_gray, -1, horizontal_sobel )

plt.imshow(filtered_image, cmap='gray')
```
